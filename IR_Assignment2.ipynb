```python
def read_file(file_name):
  file_path='/content/drive/MyDrive/Information Retrieval/Corpus-20230203T210935Z-001/Corpus/'+file_name
  try:
    with open(file_path,'r') as file:
      return file.read()
  except FileNotFoundError:
    print(f"File Not found at {file_path}")
    return None
  except Exception as e:
    print(f"An error occured while reading the file {e}")
    return None
lines = read_file("levis.txt")
```

```python
#case lowering
def lower_case(lines):
  return lines.lower()

# #removing punctuation
def remove_punct(text):
  text = "".join([char for char in text if char not in string.punctuation])
  text = re.sub('[0-9]+','', text)
  return text

def tokenization(text):
  text = text.strip()
  text = re.split('\W+', text)
  return text

#removing all 1 letter words like 's', 'A', etc
def remove_singles(text):
  for word in text:
    if len(word) < 2:
      text.remove(word)
  return text

stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  for word in text:
    if word in stopwords:
      text.remove(word)
  return text

def stemming(text):
  return [nltk.PorterStemmer().stem(word) for word in text]

def lemmatizer(text):
  return [nltk.WordNetLemmatizer().lemmatize(word) for word in text]

def processing_steps(lines):
  lines = lower_case(lines)
  lines = remove_punct(lines)
  tokens = tokenization(lines)
  tokens = remove_singles(tokens)
  tokens = remove_stopwords(tokens)
  tokens = stemming(tokens)
  tokens = lemmatizer(tokens)
  return tokens

lines_list = processing_steps(lines)
print(lines_list)
```

```python
def log_term_frequency(list_files):
  doc_frequency={}
  for i in range(len(list_files)): #every doc 
    lines = read_file(list_files[i])
    tokens = processing_steps(lines)

    for word in tokens:
      if word in doc_frequency and [i,0] not in doc_frequency[word]:
        doc_frequency[word].append([i,0])
      elif word not in doc_frequency.keys():
        doc_frequency[word] = [[i,0]]
    
  for word in doc_frequency:
    for i in range(len(doc_frequency[word])):
      lines = read_file(file_list[doc_frequency[word][i][0]])
      tokens = processing_steps(lines)

      for j in range(len(tokens)):
        if tokens[j] == word:
          doc_frequency[word][i][1] += 1 
      doc_frequency[word][i][1]=1+(math.log10(doc_frequency[word][i][1]))
  return doc_frequency

doc_f = log_term_frequency(file_list)
print(doc_f)
```

```python
from collections import defaultdict

doc_dict = defaultdict(lambda: {})
norm={}
for key, value in doc_f.items():
  for item in value:
    doc_dict[item[0]][key] = item[1]
    # norm[item[0]][key]+=item[1]

print(doc_dict)
# print(norm)
```

```python
for i in range(len(doc_dict)):
  count=0
  for key, value in doc_dict[i].items():
    count+=value*value
  norm[i]=math.sqrt(count)
  
print(norm)
```

```python
#storing normalized values in defaultdict
for doc in range(len(doc_dict)):
  for term in doc_dict[doc]:
    doc_dict[doc][term] = doc_dict[doc][term]/norm[doc]

  
print(doc_dict)
```
